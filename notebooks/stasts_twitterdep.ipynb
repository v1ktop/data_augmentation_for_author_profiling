```python
man_dir="D:/corpus/depression_twit/"
```

```python
docs_dir="all_data/"
truth_name="golden_training.txt"
```

```python
ids=[]
labels=[]
```

```python
import os
```

```python
separator="\t"
file = open(os.path.join(man_dir,truth_name), encoding="utf8")
text=file.readlines()
```

```python
for line in text:
    line = line.rstrip('\n')
    row=line.split(separator)
    ids.append(row[0])
    labels.append(row[1])
```

```python
docs=[]
```

```python
for filename in ids:
        file = open(os.path.join(man_dir+docs_dir,filename+".txt"), encoding="utf8")
        docs.append(file.read())
        file.close()
```
